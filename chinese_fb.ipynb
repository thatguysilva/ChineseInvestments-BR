```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Example time series data
data = {'ds': pd.to_datetime(['2010', '2011', '2012', '2013', '2014', '2015','2016','2017','2018','2019','2020','2021']),
        'y': [13610.32, 7299.62, 1845.04, 3055.64, 3940.55, 8124.31, 4395.76, 17638.56, 1828.70, 9180.23, 2032.70, 2652.00]}
df = pd.DataFrame(data)


# Initialize and fit the model
model = Prophet()
model.fit(df)

# Create future dates to forecast
future = model.make_future_dataframe(periods=2, freq='Y')  # Forecasting 12 months ahead

# Make predictions
forecast = model.predict(future)

# Visualize the forecast
fig, ax = plt.subplots(figsize=(10, 6))
model.plot(forecast, ax=ax)
plt.title('Forecast with Prophet', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.show()
```
